# Постановка задач
<a name = 'Постановка_задач'>

> ## Задание
Создать веб-сайт используя HTML и CSS.



##Цель работы

> Блок с отступами


<a name = 'Цель_работы'></a>

Познакомиться с редактором кода, овладеть навыками по созданию веб-страницы используя HTML.


##Задачи работы

<a name = 'Задачи_Работы'></a>

1. Установить среду Visual Studio Code и плагин для проверки результата отображения HTML файла в браузере

1. Подготовить 4 html страницы – 1-я главная (index) и 3 дочерних. Все страницы разделить на Шапку, Панель навигации, Подвал, Содержимое страницы, Боковую панель.

1. Создать на главной странице Заголовок, Описание, Ссылки на все дочерние страницы. Создать на каждой из дочерних страниц ссылки для перехода на главную страницу.
1. На одной из дочерних страниц разместить несколько фотографий
1. На следующей из дочерних страниц разместить элемент с воспроизведением видео файла
1. На последней из дочерних страниц разместить элемент с воспроизведением аудио файла
1. Добавить иконку к сайту
1. Создать таблицу стилей CSS и подключите её к созданным страницам
